Correlación entre variables

```python
# Solo columnas numéricas
corr_columns = ['vote_count', 'budget', 'revenue', 'runtime', 'vote_average', 'popularity']

# Convertir a float si alguna está como object
for col in corr_columns:
    movies_metadata[col] = pd.to_numeric(movies_metadata[col], errors='coerce')

# Calcular la matriz de correlación
correlation_matrix = movies_metadata[corr_columns].corr()
print(correlation_matrix)

# Opcional: visualizar con heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlación entre vote_count y variables numéricas")
plt.show()
```

revenue (0.81) y budget (0.68) tienen una alta correlación positiva con vote_count. Esto indica que las películas con mayor presupuesto y mayor recaudación tienden a recibir más votos.

popularity (0.56) también muestra una correlación moderada, lo que tiene sentido: películas más populares reciben más votos.

runtime (0.11) y vote_average (0.12) tienen correlaciones bajas, lo que indica que la duración y la calificación promedio influyen poco en la cantidad de votos.

```python
import ast

def get_main_genre(genres):
    try:
        g = ast.literal_eval(genres)
        if g: return g[0]['name']
    except: return None

movies_metadata['main_genre'] = movies_metadata['genres'].apply(get_main_genre)
plt.figure(figsize=(12,5))
sns.countplot(data=movies_metadata, x='main_genre', order=movies_metadata['main_genre'].value_counts().index, palette='Set2')
plt.xticks(rotation=45)
plt.title("Número de películas por género")
plt.show()
```

analisis estadistico

La mayoría de las películas tienen pocos votos y calificaciones promedio entre 5 y 7. Los géneros más frecuentes son Drama, Comedy y Action, mientras que Documental y Horror son menos comunes. Presupuesto y recaudación están fuertemente correlacionados: películas con mayor inversión suelen generar más ingresos y votos. La duración típica es de 80 a 110 minutos, y solo unas pocas películas destacan por alta popularidad o valores extremos en presupuesto y duración.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar los datos por género principal y sumar la cantidad de votos
votes_por_genero = movies_metadata.groupby('main_genre')['vote_count'].sum().reset_index()

# Ordenar de mayor a menor
votes_por_genero = votes_por_genero.sort_values(by='vote_count', ascending=False)

# Crear gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(data=votes_por_genero, x='vote_count', y='main_genre', palette='viridis')
plt.title('Cantidad de votos por género', fontsize=16)
plt.xlabel('Cantidad de votos', fontsize=12)
plt.ylabel('Género', fontsize=12)
plt.show()
```

El gráfico muestra que algunos géneros como Drama, Comedy y Action concentran la mayoría de los votos, mientras que géneros como Documental u Horror tienen pocos votos

```python
# Histograma de duración de películas con ticks cada 20 minutos
plt.figure(figsize=(12, 6))
plt.hist(movies_metadata['runtime'].dropna(), bins=50, color='coral', edgecolor='black')
plt.title('Distribución de la Duración de las Películas', fontsize=16)
plt.xlabel('Duración (minutos)', fontsize=12)
plt.ylabel('Cantidad de Películas', fontsize=12)

# Ajustar los ticks cada 20 minutos
max_runtime = int(movies_metadata['runtime'].dropna().max())
plt.xticks(range(0, max_runtime + 20, 20), rotation=45)

plt.show()

```

La mayoría de las películas en el dataset tienen una duración entre 80 y 100 minutos
